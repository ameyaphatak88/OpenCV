```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

image1 = cv2.imread('ameya.jpeg')
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)

plt.imshow(image1)
```

```python
gray_img = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_img)
```

```python
hsv_img = cv2.cvtColor(image1, cv2.COLOR_BGR2HSV)
```

```python
plt.imshow(hsv_img)
```

```python
ret,thresh_binary = cv2.threshold(image1,127,255,cv2.THRESH_BINARY)
```

```python
plt.imshow(thresh_binary, 'gray')
```

```python
ret,thresh_trunc = cv2.threshold(image1,200,255,cv2.THRESH_TRUNC)
```

```python
plt.imshow(thresh_trunc, 'gray')
```

```python
edges = cv2.Canny(image1,100,200) 
```

```python
plt.imshow(edges)
```
